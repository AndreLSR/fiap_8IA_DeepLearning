## 2) Importing and separating train, validation and test datasets

```python
batch_size = 128
```

```python
train_datagen = preprocessing.image.ImageDataGenerator(validation_split=0.20)
train_generator = train_datagen.flow_from_directory(
    'Dataset_Train',
    batch_size=batch_size,
    color_mode='grayscale',
    seed=42,
    class_mode="categorical",
    shuffle=True,
    target_size=(28, 28)
)
```

```python
train_generator
```

```python
valid_datagen = preprocessing.image.ImageDataGenerator(validation_split=0.10)
validation_generator = valid_datagen.flow_from_directory(
    'Dataset_Train',
    batch_size=batch_size,
    color_mode='grayscale',
    seed=42,
    class_mode="categorical",
    shuffle=True,
    target_size=(28, 28),
)
```

```python
test_datagen = preprocessing.image.ImageDataGenerator()
test_generator = test_datagen.flow_from_directory(
    'Dataset_Test',
    batch_size=1,
    color_mode = 'grayscale',
    seed=42,
    class_mode=None,
    shuffle=False,
    target_size=(28, 28)
)
```

```python
num_classes = train_generator.num_classes
num_classes
```

## 4) Building the LeNet-5 Model

![](https://i.ibb.co/L6Tr8s0/LeNet5.png)

```python
model = Sequential()
```

### a) Convolução 2D com função de ativação tanh 6 filtros de 5x5

```python
model.add(Conv2D(6, (5,5), input_shape=(28,28,1), padding='same', activation='tanh'))
print(model.output_shape)
```

#### b) Camada de Pooling

```python
model.add(AveragePooling2D(pool_size=(2,2), strides=(1,1), padding='valid'))
```

#### c) Convolução 2D com função de ativação tanh 16 filtros de 5x5

```python
model.add(Conv2D(16,(5,5), strides=(1,1), padding='valid', activation='tanh'))
print(model.output_shape)
```

#### d) Camada de Pooling

```python
model.add(AveragePooling2D(pool_size=(2,2), strides=(2,2), padding='valid'))
```

#### e) Convolução 2D com função de ativação tanh 120 filtros de 5x5

```python
model.add(Conv2D(120,(5, 5), strides=(1, 1), activation='tanh', padding='valid'))
print(model.output_shape)
```

#### e) Converte o conjunto de imagens num vetor unidimensional para a entrada da Camada Completamente Conectada

```python
model.add(Flatten())
```

#### f) Adicionamos uma camada completamente conectada, baseada na LeNet5 que começa com mais neurônios na camada inicial e ela vai reduzindo.

```python
model.add(Dense(84, activation='tanh'))
print(model.output_shape)
```

```python
model.add(Dense(num_classes, activation='softmax'))
print(model.output_shape)
```

### Visualizing the Model

```python
model.summary()
```

### Fitting the Model

#### Compila o modelo definindo o SGD optimization como algoritmo de atualização

```python
model.compile(
    loss='categorical_crossentropy'
    ,optimizer='sgd'
    ,metrics=['accuracy']
#     ,mode=['min']
)
```

```python
log_path = '/Users/andrerodrigues/Google Drive/FIAP-MBA 8IA/009-Deep Learning & Reinforcement Learning/Exercicios/Exercicio8_fruits15'
```

```python
tensorboard = tb(
    log_dir=log_path
    ,write_graph=False #This eats a lot of space. Enable with caution!
    #,histogram_freq = 1
    ,write_images=True
    ,batch_size = batch_size
    ,write_grads=True
)
```

```python
earlystopper = EarlyStopping(
    monitor='loss'
    ,patience=15
    ,verbose=1
)
```

```python
checkpointer = ModelCheckpoint(
    'Fruits15.hdf'
    , verbose = 1
    , save_best_only = True
    , monitor='accuracy'
    , mode='min'
)
```

```python
callbacks = [checkpointer, earlystopper, tensorboard]
```

```python
train_generator.n
```

```python
train_generator.batch_size
```

```python
validation_generator.n
```

```python
validation_generator.batch_size
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_TRAIN
```

```python
STEP_SIZE_VALID=validation_generator.n//validation_generator.batch_size
STEP_SIZE_VALID
```

```python
epochs = 50
```

```python
# fit model
history = model.fit_generator(
    generator=train_generator
    ,steps_per_epoch=STEP_SIZE_TRAIN
    ,validation_data=validation_generator
    ,validation_steps=STEP_SIZE_VALID
    ,epochs=epochs
    ,callbacks= callbacks
#     , class_weight = class_weight
)
```

### Evaluating the Model

```python
scores = model.evaluate_generator(generator=validation_generator, steps=STEP_SIZE_VALID)
```

##### The fit_generator() function called to fit the model returns a dictionary containing the loss and F-beta scores recorded each epoch on the train and test dataset. We can create a plot of these traces that can provide insight into the learning dynamics of the model.

### Learning Curves

#### Plot Training & Validation accuracy values

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()
```

#### Plot Training & Validation losses values

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Entropy Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['loss', 'val loss'], loc='lower right')
plt.show()
```

### Predicting the output

```python
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size

test_generator.reset()

pred=model.predict_generator(test_generator,
                             steps=STEP_SIZE_TEST,
                             verbose=1
                            )
```

```python
pred
```

```python
pred.shape
```

```python
predicted_class_indices=np.argmax(pred,axis=1) # Returns the indices of the maximum values along an axis.
predicted_class_indices
```
