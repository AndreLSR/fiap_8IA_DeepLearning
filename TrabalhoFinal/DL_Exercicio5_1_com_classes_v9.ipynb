## 3)Validando independencia entre as variáveis

### a) correlation

```python
set_option('precision', 2)
corr = df.corr(method='pearson')
print(corr)
plot1 = sb.heatmap(df.corr()) 
print('\n')
plt.title("Heatmap - Check de Indepedencia das Variaveis")
plt.show(plot1)
```

## 4) Separando os dados de teste e treinamento

```python
X = df.iloc[:,0:17]
X
```

```python
inputdim=len(X.columns)
inputdim
```

```python
y = df.iloc[:,17:]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)
```

## 5) Preparando o modelo de MLP

### a) Aplicando o modelo Sequencial

```python
#quantidade de Neuronios semelhantes ao approach de Deepa, 2013
#4n¨2

model = tf.keras.Sequential()
model.add(layers.Dense(800, input_dim=inputdim, activation='relu'))
model.add(layers.Dense(400, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(num_clusters, activation='softmax'))
```

### b) Visualizando o modelo

```python
print(model.summary())
```

```python
plot_model(model, to_file='cnn-ex5_1.png', show_shapes=True, show_layer_names=True)
Image('cnn-ex5_1.png')
```

### c) Salvando o checkpoint

```python
# mcp_save = ModelCheckpoint('Ex5_1_AR_v1.hdf', save_best_only=True, monitor='val_accuracy', mode='max', verbose=1)
```

### d) Compilando o Modelo

```python
batch_size = 5
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#opt = Adamax(learning_rate=0.002, beta_1=0.9, beta_2=0.999)
#model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
log_path = '../Exercicio5.1_com_classes'
```

```python
tensorboard = tb(
    log_dir=log_path
    ,write_graph=False #This eats a lot of space. Enable with caution!
    #,histogram_freq = 1
    ,write_images=True
    ,batch_size = batch_size
    ,write_grads=True
)
```

```python
earlystopper = EarlyStopping(
    monitor='loss'
    ,patience=15
    ,verbose=1
)
```

```python
checkpointer = ModelCheckpoint(
    'Ex5_1_AR_v1.hdf'
    , verbose = 1
    , save_best_only = True
    , monitor='accuracy'
    , mode='min'
)
```

```python
callbacks = [checkpointer, earlystopper, tensorboard]
```

### e) Treinando o Modelo

```python
history  = model.fit(
    X_train
    ,y_train
    ,validation_split = 0.1
    ,epochs=50
#     ,batch_size=5
    ,callbacks=callbacks
)
```

### f) Saving the Model

```python
model.save('../Exercicio5.1_com_classes/variables')
```

### g) Verificando acurácia alcançada

```python
scores = model.evaluate(X_test, y_test)
```

```python
print("\n%s: %.2f%%" % (model.metrics_names[0], scores[0]*100))
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Entropy Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['loss', 'val loss'], loc='lower right')
plt.show()
```

## 6) Corrigindo o modelo de MLP

#### Para corrigir o overfit vamos mudar o optimizer (aka Algoritmo de Descida do Gradiente) de adam para sgd (Stochastic gradient descent).

#### A Descida do Gradiente é um algoritmo de otimização usado para encontrar os valores de parâmetros (coeficientes ou se preferir w e b – weight e bias) de uma função que minimizam uma função de custo. A Descida do Gradiente é melhor usada quando os parâmetros não podem ser calculados analiticamente (por exemplo, usando álgebra linear) e devem ser pesquisados por um algoritmo de otimização.

# 7) Corrigindo o modelo de MLP usando Weight Regulation

#### A regularização de peso(Weight regulation) fornece uma abordagem para reduzir o overfitting de um modelo de rede neural de aprendizado profundo nos dados de treinamento e melhorar o desempenho do modelo em novos dados, como o conjunto de testes de validação.

```python
from keras.regularizers import l2
```

### a) Aplicando o modelo Sequencial

```python
#quantidade de Neuronios semelhantes ao approach de Deepa, 2013
#4n¨2

# # Acc = 94.8%
# model3 = Sequential()
# model3.add(Dense(256, input_dim=17, activation='relu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dropout(0.5))
# model3.add(Dense(128, activation='relu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dropout(0.5))
# model3.add(Dense(64, activation='relu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dense(num_clusters, activation='sigmoid'))

# # # Acc = 0%
# model3 = Sequential()
# model3.add(Dense(256, input_dim=17, activation='elu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dropout(0.5))
# model3.add(Dense(128, activation='elu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dropout(0.5))
# model3.add(Dense(64, activation='elu', kernel_regularizer=l2(int(0.001))))
# model3.add(Dense(num_clusters, activation='softmax'))

# Acc = 94.72%
# model3 = Sequential()
# model3.add(Dense(256, input_dim=17, activation='relu', kernel_regularizer=l2(int(0.01))))
# model3.add(Dropout(0.5))
# model3.add(Dense(128, activation='relu', kernel_regularizer=l2(int(0.01))))
# model3.add(Dropout(0.5))
# model3.add(Dense(64, activation='relu', kernel_regularizer=l2(int(0.01))))
# model3.add(Dense(num_clusters, activation='sigmoid'))

# Acc = 95%
model3 = Sequential()
model3.add(Dense(256, input_dim=17, activation='relu', kernel_regularizer=l2(int(0.0001))))
model3.add(Dropout(0.5))
model3.add(Dense(128, activation='relu', kernel_regularizer=l2(int(0.00001))))
model3.add(Dropout(0.5))
model3.add(Dense(64, activation='relu', kernel_regularizer=l2(int(0.00001))))
model3.add(Dense(num_clusters, activation='sigmoid'))
```

### b) Visualizando o modelo

```python
print(model3.summary())
```

```python
plot_model(model3, to_file='cnn-ex5_1_3.png', show_shapes=True, show_layer_names=True)
Image('cnn-ex5_1_3.png')
```
